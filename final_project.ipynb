```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
carbon_df = pd.read_csv("CarbonDataEmissions.csv")
transportation_df = pd.read_csv("TransportationData.csv")
```

```python
carbon_df.rename(columns={"Percent": "percent_change", "Absolute": "absolute_change"}, inplace=True)
carbon_df.head()
```

```python
transportation_df.rename(columns={"Commute mode share (percent)": "commute_mode_share_percent"}, inplace=True)
transportation_df["commute_mode_share_percent"] = transportation_df["commute_mode_share_percent"] * 100
transportation_df.head()
```

```python
alt.renderers.enable('default')
alt.data_transformers.enable('default', max_rows=None)
```

```python
alt.Chart(carbon_df).mark_bar().encode(
    alt.X('State'),
    alt.Y('percent_change'),
    color=alt.Color('percent_change', scale=alt.Scale(scheme='accent'), sort=alt.Sort()),
).properties(
    title='Carbon Emissions Percent Change 2000-2018'
)
```
